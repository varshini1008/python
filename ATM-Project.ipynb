```python
class ATM():
    def __init__(self):
        self.balance=0
        self.__pin=0
        self.info()

    
    def info(self):
        input_value=int(input(""""1.create the pin for user\n
                    2.check the balance\n
                    3.withdrawal\n
                    4.deposit\n
                    5.mini statement\n
                    6.change the pin\n
                    7.Exit"""))
        
        def change_pin():
            check=int(input("enter the pin"))
            if check==self.__pin:
                new_pin1=int(input("enter the 4 digit pin"))
                new_pin2=int(input("enter the 4 digit pin"))
                if new_pin1==new_pin2:
                    print("pin changed successful")
                    self.__pin=new_pin1
                else:
                    print("pin mismatch") 
                    
            if input("type b to go back to main menu or e to exit")=="b":
                self.info()
            else:
                pass
        
        def create_pin():
            new_pin1=int(input("enter the 4 digit pin"))
            new_pin2=int(input("enter the 4 digit pin"))
            if new_pin1==new_pin2:
                print("pin successful")
                self.__pin=new_pin1
            else:
                print("pin mismatch")
                
            if input("type b to go back to main menu or e to exit")=="b":
                self.info()
            else:
                pass
            
        def balance():
            check=int(input("enter the pin"))
            if check==self.__pin:
                print("enter the balance is {}".format(self.balance))
            else:
                print("pin mismatch")
        
            if input("type b to go back to main menu or e to exit")=="b":
                self.info()
            else:
                pass
           
            
            
        def deposit():
            check=int(input("enter the pin"))
            if check==self.__pin:
                how_much=int(input("enter how much amount do you want to deposit"))
                notes_100=int(input("how many 100 notes"))
                notes_200=int(input("how many 200 notes"))
                notes_500=int(input("how many 500 notes"))
                notes_2000=int(input("how many 2000 notes"))
                if how_much==notes_100*100+notes_200*200+notes_500*500+notes_2000*2000:
                    self.balance=how_much+self.balance
                    print("despoited successfully")
                else:
                    print("notes mismatch")
        
        def withdrawal():
            check=int(input("enter the pin"))
            if check==self.__pin:
                how_much_withdrawal=int(input("enter the amount you want to withdaw"))
                if how_much_withdrawal<=self.balance:
                    self.balance=self.balance-how_much_withdrawal
                    
                    print("withdrawal successful")
                else:
                    print("insufficient amount")
                
        if input_value==1:
            create_pin()
        elif input_value==2:
            balance()
        elif input_value==3:
            withdrawal()
        elif input_value==4:
            deposit()
        elif input_value==6:
            change_pin()
        else:
            pass
      
```
